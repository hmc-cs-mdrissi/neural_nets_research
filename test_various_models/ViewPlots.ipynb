```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
plot_every = 100
suffix = "accuracy.txt"
matching_files = []
files = [f for f in os.listdir('.') if os.path.isfile(f)]
for name in files:
    if name[-len(suffix):] == suffix:
        matching_files.append(name[:-len(suffix)])
print(matching_files)

for filename in matching_files:

    with open(filename + "loss.txt") as f:
        print(filename)
#         print(f.read().split(","))
        points = [float(x) for x in f.read().split(",")[:-1]]

    plt.plot([x * plot_every for x in range(len(points))], points)
    plt.title(filename + "loss")
    plt.show()

    with open(filename + "accuracy.txt") as f:
        points = [float(x) for x in f.read().split(",")[:-1]]

    plt.plot([x * plot_every for x in range(len(points))], points)
    plt.title(filename + "accuracy")
    plt.show()
```
